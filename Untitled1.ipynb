```python
# Step 1: Upload file
from google.colab import files
uploaded = files.upload()

# Step 2: Import Pandas
import pandas as pd

# Step 3: Read the Excel/CSV file
data = pd.read_excel("students_50.xlsx")   # or pd.read_csv("students_50.csv")

# Step 4: Show first 5 rows
print("üìä First 5 rows:")
print(data.head())

# Step 5: Average marks per subject
print("\nüìå Average Marks per Subject:")
print(data[['Math', 'Science', 'English']].mean())

# Step 6: Topper in each subject
print("\nüèÜ Subject Toppers:")
for subject in ['Math', 'Science', 'English']:
    topper = data.loc[data[subject].idxmax()]
    print(f"{subject} Topper: {topper['Name']} ({topper[subject]} marks)")

# Step 7: Overall topper (based on total marks)
data['Total'] = data[['Math', 'Science', 'English']].sum(axis=1)
overall_topper = data.loc[data['Total'].idxmax()]
print(f"\nüéØ Overall Topper: {overall_topper['Name']} (Total {overall_topper['Total']} marks)")

# Step 8: Pass/Fail count (Pass = marks >= 40 in all subjects)
data['Pass'] = (data[['Math', 'Science', 'English']] >= 40).all(axis=1)
pass_count = data['Pass'].sum()
fail_count = len(data) - pass_count
print(f"\n‚úî Passed: {pass_count}, ‚ùå Failed: {fail_count}")

# Step 9: Visualization (Average per subject)
import matplotlib.pyplot as plt

data[['Math','Science','English']].mean().plot(kind='bar', color=['skyblue','lightgreen','salmon'])
plt.title("Average Marks per Subject")
plt.ylabel("Marks")
plt.show()
```
